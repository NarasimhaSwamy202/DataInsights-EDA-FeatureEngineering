## Fetaure Engineering Process

```python
# final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0].astype(int)
# final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1].astype(int)
# final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2].astype(int)
# ----------------------------------OR--------------------------------------------
final_df["Date"]=final_df['Date_of_Journey'].apply(lambda x:x.split("/")[0]).astype(int)
final_df["Month"]=final_df['Date_of_Journey'].apply(lambda x:x.split("/")[1]).astype(int)
final_df["Year"]=final_df['Date_of_Journey'].apply(lambda x:x.split("/")[2]).astype(int)
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df.info()
```

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
# Handling Total_stops column
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df.Total_Stops[(final_df['Source'] == 'Delhi') & (final_df['Destination'] == 'Cochin') & (final_df['Airline'] == 'Air India')].value_counts()
```

```python
mapping_dict = {'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan: 2}
final_df['Total_Stops'] = final_df['Total_Stops'].map(mapping_dict)
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

"We don't need the 'Route' column because the 'Source' and 'Destination' columns provide the same information."

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['duration_hour'] = final_df['Duration'].str.split('h').str[0]
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[-1].str.split('m').str[0]
```

```python
# final_df['duration_hour']=final_df['duration_hour'].astype('int')
# final_df['duration_min']=final_df['duration_min'].astype('int')
```

The above code gives the errors because some of the records contains only hours or mins

```python
final_df[final_df['duration_hour']=='5m']
```

"It is highly improbable for a journey from Mumbai to Hyderabad with 2 total stops to have a duration of just 5 minutes."

So let us drop these two records

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
# Extracting minutes if 'm' is present, otherwise setting it to 0
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[-1]
final_df['duration_min'] = final_df['duration_min'].str.split('m').str[0]
final_df['duration_min'] = final_df['duration_min'].apply(lambda x: 0 if 'h' in x else x)
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
# Converting duration_hour and duration_min to integer datatype
final_df['duration_hour'] = final_df['duration_hour'].astype(int)
final_df['duration_min'] = final_df['duration_min'].astype(int)
```

```python
final_df.head(1)
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
